```python
import numpy as np
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

```python

import io
df=pd.read_csv(io.BytesIO(uploaded['StudentsPerformance.csv']))
df
```

```python
# Find out how many males and females participated in the test.

print(df['gender'].value_counts())
```

```python
# What do you think about the students' parental level of education?

print(df['parental level of education'].value_counts())
```

```python
# Who scores the most on average for math, reading and writing based on gender and test prearation course

df['average score'] = (df['math score'] + df['writing score'] + df['reading score']) / 3
df
```

```python
df1 = df.groupby(['gender','parental level of education'])[['reading score', 'math score', 'writing score']].max()
df1
```

```python
# What do you think about the scoring variation for math, reading and writing based on Gender, Test preparation course

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize= (7,7))
plt.title('Gender v/s average score')
sns.barplot(x = 'gender', 
            y = 'average score',
            data = df);
plt.figure(figsize= (7,7))
plt.title('Test Preparation course v/s average score')
sns.barplot(x = 'test preparation course', 
            y = 'average score',
            data = df);
     
```
